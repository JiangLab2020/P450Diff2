```python
import numpy as np
import pandas as pd
```

```python
Blast_path = "data/compare/similar/p450diff2.tsv"
pLDDT_path = "data/compare/plddt/P450Diff2"
```

```python
dfBlast = pd.read_csv(
    Blast_path,
    sep="\t",
    header=None,
    names=[
        "qseqid",
        "sseqid",
        "pident",
        "length",
        "mismatch",
        "gapopen",
        "qstart",
        "qend",
        "sstart",
        "send",
        "evalue",
        "bitscore",
    ],
)
```

```python
dfBlast
```

```python
dfPlddt1 = pd.read_csv(pLDDT_path + "/1/results.txt", sep="\t")
dfPlddt1["Sequence ID"] = "generated_samples_string_0_" + dfPlddt1["Sequence ID"]
dfPlddt2 = pd.read_csv(pLDDT_path + "/2/results.txt", sep="\t")
dfPlddt2["Sequence ID"] = "generated_samples_string_1_" + dfPlddt2["Sequence ID"]
dfPlddt3 = pd.read_csv(pLDDT_path + "/3/results.txt", sep="\t")
dfPlddt3["Sequence ID"] = "generated_samples_string_2_" + dfPlddt3["Sequence ID"]
dfPlddt4 = pd.read_csv(pLDDT_path + "/4/results.txt", sep="\t")
dfPlddt4["Sequence ID"] = "generated_samples_string_3_" + dfPlddt4["Sequence ID"]
dfPlddt5 = pd.read_csv(pLDDT_path + "/5/results.txt", sep="\t")
dfPlddt5["Sequence ID"] = "generated_samples_string_4_" + dfPlddt5["Sequence ID"]
dfPlddt6 = pd.read_csv(pLDDT_path + "/6/results.txt", sep="\t")
dfPlddt6["Sequence ID"] = "generated_samples_string_5_" + dfPlddt6["Sequence ID"]
dfPlddt7 = pd.read_csv(pLDDT_path + "/7/results.txt", sep="\t")
dfPlddt7["Sequence ID"] = "generated_samples_string_6_" + dfPlddt7["Sequence ID"]
dfPlddt8 = pd.read_csv(pLDDT_path + "/8/results.txt", sep="\t")
dfPlddt8["Sequence ID"] = "generated_samples_string_7_" + dfPlddt8["Sequence ID"]

dfPlddt = pd.concat(
    [dfPlddt1, dfPlddt2, dfPlddt3, dfPlddt4, dfPlddt5, dfPlddt6, dfPlddt7, dfPlddt8],
    ignore_index=True,
)
```

```python
dfcombined = pd.merge(
    dfBlast, dfPlddt, left_on="qseqid", right_on="Sequence ID", how="inner"
)
```

```python
dfcombined = dfcombined[dfcombined[" pLDDT"] > 70]
```
